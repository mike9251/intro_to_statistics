## 2.5 Множественные сравнения в ANOVA

При множественном сравнении всех возможных пар p-value увеличивается, т.е. растет вероятность ошибки первого рода - отклонить корректную нулевую гипотезу. Т.е. с ростом числа групп из одной ГС будет расти вероятность отклонить правильную нулевую гипотезу, хотя ни каких стат различий между ними нет.

Пример: при однократном подбрасывании честной монеты, вероятность выпадения решки 0.5. При 1000 подбрасываниях, вероятность того, что хотя бы один раз выпадет решка -> 1.

Чтобы скоректировать рост вероятности ошибки первого рода используют поправки на множественное сравнение.

### 1. Поправка Бенферони
Уменьшаем порог p-value обратно пропорционально числу попарных сравнений.

p' = p / (N*(N-1)/2), N - число групп.

+ простота
- растет вероятность пропустить стат различия (пропустить открытие). На практике число групп очень большое и получить статистически значимые различия становится невозможным из-за очень маленького порога p-value.

### 2. Поправка Тьюки
Попарно расчитывается t-test но немного по-другому вычисляется стандартеая ошибка se.
Позволяет расчитать доверительный интервал разницы между средними значениями. Если такой интервал не включает 0, тогда можно отклонить Н0 гипотезу (о равенстве средних). Пример, для пар C-A, D-A и D-B есть статистически значимые различия. Для пары C-B мы не отклоняем Н0, т.к. p-value выше порога. Такой результат получился как раз за счет применения поправки. Если бы просто сделали t-test то значение было бы 0.04 и Н0 была бы отклонена.
